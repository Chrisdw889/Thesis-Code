```python
import pickle
import os

import StaticThreshold as sth

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

# Load in Sim Output

```python
directory = 'INSERT DIRECTORY' # Directory where events generated by Simulation.py are saved.
    
background_file = '0_SNR_data.pkl' # Baseline eventless data filename.

files = [] # List of simulation filenames

SNRs = [float(x.split('_')[0]) for x in files]

# Load Simulated Traces
raw_traces, ground_truths = sth.load_SNR_data(directory, files)

st_devs = [1, 2, 3, 4, 5] # Event detection sensitivities

# Load background trace
with open(os.path.join(directory, background_file), 'rb') as f:
    background_data = pickle.load(f)[0]
        
```

# Event Search
Perform event searches with all combinations of SNRs and sensitivities

```python
columns = ['SNR', 'Truths', 'St_Dev', 'Predictions', 'event_locations']
df = pd.DataFrame(columns=columns)

for snr in SNRs:
    for std in st_devs:
        print("SNR: " + str(snr))
        print("STD: " + str(std))
        traces = raw_traces[snr]
        truths = ground_truths[snr]
        
        evt_locs = sth.event_search(background_data, traces, std, snr)
        print("Event Search Finished")
        predictions = sth.locs_to_preds(evt_locs, traces)
        
        df2 = pd.DataFrame([[snr, truths, std, predictions, evt_locs]], columns=columns)
        df = df.append(df2)
```

# Save Event Locations

```python
df_temp = df.set_index(['SNR', 'St_Dev'])
evt_locs = df_temp.event_locations

directory = 'INSERT DIRECTORY' # Save directory
fname = os.path.join(directory, 'INSERT FILENAME') # Save filename
with open(fname, 'wb') as f:
    pickle.dump(evt_locs, f)
```
