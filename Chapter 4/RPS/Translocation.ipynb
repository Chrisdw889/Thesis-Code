```python
import os

import numpy as np

from scipy.io import loadmat

import matplotlib.pyplot as plt
```

# Load Data

```python
data = loadmat("dataout_333.mat.cor.mat")

trace = data['ch2']
trace.shape
```

# Plot example RPS event¶¶

```python
fig, ax = plt.subplots(figsize=(8, 6))

x = np.linspace(0, len(trace) / 1E6, len(trace))

start = 6677000
end = 6681000

ax.plot(x[start:end], trace[start:end])

ax.set_ylabel("Current / A", weight='bold')
ax.set_xlabel("Time / s", weight='bold')

#ax.set(ylim=(-3E-11, 7E-11))

ax.tick_params(width=1.5)
for axis in ['top', 'right', 'bottom', 'left']:
    ax.spines[axis].set_linewidth(1.5)
```

```python
fig.savefig("Example Transloc.png", dpi=600)
```
